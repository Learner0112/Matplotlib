# Matplotlib Example Workflow

```python

# 1. Prepare the data
x = [1,2,3,4]
y= [11,22,33,44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (12,5))
                       
# 3. Plot Data
ax.plot(x,y)

# 4. Customize Plot 
ax.set(title = "Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & Show (save the whole figure)
fig.savefig("images/sample_plot.png")
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1,ax2),(ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2,figsize = (10,5))

ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
# Subplots Option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

#Plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

# Pyplot vs matplotlib oo method
# pyplot for quick visua for 

```python
over_50 = heart_disease[heart_disease.age >50]
over_50
```

```python
#pyplot method
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            figsize = (10,6),
            );
```

```python
# OO Method mixed with pyplot method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            ax =  ax);
#ax.set_xlim([45,100])
#ax.set_ylim([])
```

```python
# OO method from scratch
fix, ax = plt.subplots (figsize = (10,6))

#plot the data
scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'])

#customize the plot
ax.set(title =  'This is title',
      xlabel = 'Age',
      ylabel = 'Cholesterol')

#set legend
ax.legend(*scatter.legend_elements(), title = "Target")

# Add horizontal line in matplotlib plot
ax.axhline(over_50['chol'].mean(),
         linestyle = '--' )
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2,ncols = 1,figsize = (15,15), sharex = True)

#plot the data
scatter = ax0.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'])

scatter = ax1.scatter(x = over_50['age'],
                    y = over_50['thalach'],
                    c = over_50['target'])

#customize the plot
ax0.set(title =  'This is title',
      
      ylabel = 'Cholesterol')

ax1.set(title =  'This is title',
      xlabel = 'Age',
      ylabel = 'thalach')


#set legend
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add horizontal line in matplotlib plot
ax0.axhline(over_50['chol'].mean(),
         linestyle = '--' )

#set legend
ax1.legend(*scatter.legend_elements(), title = "Target")

# Add horizontal line in matplotlib plot
ax1.axhline(over_50['thalach'].mean(),
         linestyle = '--' )

# Add a title
fig.suptitle("Heart Disease Analysis", fontsize = 15);

```

```python
plt.style.use('seaborn-whitegrid')


fig, (ax0, ax1) = plt.subplots(nrows = 2,ncols = 1,figsize = (15,10), sharex = True)

#plot the data
scatter = ax0.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'],
                     cmap = 'winter')

scatter = ax1.scatter(x = over_50['age'],
                    y = over_50['thalach'],
                    c = over_50['target'],
                     cmap = 'winter') # change color 
#x axis limit
ax1.set_xlim([50,80,])
ax0.set_xlim([50,80,])

ax1.set_ylim([60,200,])
ax0.set_ylim([60,500,])


#customize the plot
ax0.set(title =  'This is title',
      
      ylabel = 'Cholesterol')

ax1.set(title =  'This is title',
      xlabel = 'Age',
      ylabel = 'thalach')


#set legend
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add horizontal line in matplotlib plot
ax0.axhline(over_50['chol'].mean(),
         linestyle = '--' )

#set legend
ax1.legend(*scatter.legend_elements(), title = "Target")

# Add horizontal line in matplotlib plot
ax1.axhline(over_50['thalach'].mean(),
         linestyle = '--' )

# Add a title
fig.suptitle("Heart Disease Analysis", fontsize = 15);


#Save the fig
fig.savefig('HDA_1.png')
```
