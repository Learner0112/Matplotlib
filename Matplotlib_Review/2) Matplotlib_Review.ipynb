```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x  =  np.linspace(0,5,11)
y = x **2
```

```python
# Functional Method
plt.plot(x,y, 'r-')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
plt.show()
```

```python
# for multiplot
plt.subplot(1,2,1)
plt.plot(x,y, 'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show();
```

```python
# Object oriented method - OO
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(x,y)
ax.set_xlabel('x lable')
ax.set_ylabel('y label')
ax.set_title('Set title')

plt.show();
```

```python
#Manual Method
fig = plt.figure()

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y)
ax2.plot(x,y)

ax2.set_title("Small")
ax1.set_title("large")
```

```python
#
fig, ax = plt.subplots(nrows = 1, ncols = 2)

#ax.plot(x,y)
#plt.tight_layout()
#iterate through
for current_ax in ax:
    current_ax.plot(x,y)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].plot(x,y)
ax[1].plot(y,x)

plt.show()
```

```python
#Figure size , aspect ratio, DPI
fig = plt.figure(figsize = (8,2))

ax = fig.add_axes ([0,0,1,1])
ax.plot(x,y)
```

```python
#Figure size , aspect ratio, DPI
fig, ax = plt.subplots(nrows =2, ncols = 1, figsize = (8,2))

ax[0].plot(x,y)
ax[1].plot(y,x)

plt.tight_layout()

plt.show();
```

```python
#Figure size , aspect ratio, DPI
fig = plt.figure(figsize = (8,2))

ax = fig.add_axes ([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
#Figure size , aspect ratio, DPI
fig = plt.figure(figsize = (8,2))

ax = fig.add_axes ([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X Cube') 

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax.legend(loc = 'best')
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'orange', linewidth = 3,alpha = 1., linestyle = ':', marker  = 'o', markersize = 20,
       markerfacecolor ='yellow',markeredgewidth = 3, markeredgecolor='blue' ) #rgb hex code 
# or ls/linestyle #lw line width

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
